## Flattening the Json data

```python
#accessing first index of genre column
#here it producing the json string
merged_df['genres'][0]
```

```python
# convert first index of json string into python dictionary
import json

json_d = json.loads(merged_df['genres'][0])
json_d
```

```python
type(json_d)
```

```python
#extracting the values from the dictionary

genre_list = []

for i in range(len(json_d)):
    genre_list.append(json_d[i]['name'])
```

```python
genre_list
```

```python
# now need to flatten the whole column
def cl_json(x):
    gr_list = []
    for i in range(len(x)):
        gr_list.append(x[i]['name'])
    return str(gr_list)
        
    
```

```python
merged_df['genres'].apply(json.loads).apply(cl_json)
```

```python
type(merged_df['genres'])
```

```python
def flat_json(column):
    merged_df[column] = merged_df[column].apply(json.loads).apply(cl_json)


flat_json('cast')
```

```python
flat_json('genres')
```

# Dropping Zero values

```python
merged_df.describe()
```

```python
# where some of the columns having minimum value as zero
#where budget and revenue affects the analysis if it has zero values
z=['budget','revenue']
merged_df[z] = merged_df[z].replace(0,np.nan)
merged_df.dropna(subset=z,inplace=True)
```
